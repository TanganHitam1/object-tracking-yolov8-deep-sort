```python
import os
import cv2
import numpy as np
import random
```

```python
video_path = os.path.join('.', 'output_video.mp4')
# video_out_path = os.path.join('.', 'outAFLink.mp4')
# input_ = np.loadtxt('./outputAFlink.txt', delimiter=',')
# video_out_path = os.path.join('.', 'outGSI.mp4')
# input_ = np.loadtxt('./outputGSI.txt', delimiter=',')
video_out_path = os.path.join('.', 'out_MOT17-02-FRCNN_AFLinkGSI.mp4')
input_ = np.loadtxt('./out_MOT17-02-FRCNN_AFlinkGSI.txt', delimiter=',')

colors = [(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for j in range(10)]
```

```python
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()

cap_out = cv2.VideoWriter(video_out_path, cv2.VideoWriter_fourcc(*'MP4V'), cap.get(cv2.CAP_PROP_FPS),
                        (frame.shape[1], frame.shape[0]))

```

```python
i = 0
while ret:
    tracked = input_[input_[:, 0] == i]
    for track in tracked:
        cv2.rectangle(frame, (int(track[2]), int(track[3])), (int(track[2]+track[4]), int(track[3]+track[5])), colors[int(track[1]) % len(colors)], 3)
        cv2.putText(frame, str(int(track[1])), (int(track[2]), int(track[3])), cv2.FONT_HERSHEY_SIMPLEX, 1, colors[int(track[1]) % len(colors)], 2, cv2.LINE_AA)
    print(tracked)
    # cv2.imshow('frame', frame)
    i+=1
    cap_out.write(frame)
    # if cv2.waitKey(10) & 0xFF == ord('q'):
    #     break
    ret, frame = cap.read()
cap.release()
cap_out.release()
# if cv2.waitKey(0) & 0xFF == ord('q'):
cv2.destroyAllWindows()
```
